```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE'

import cv2

# make inferences
from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor
from detectron2 import model_zoo # img lib

'''Custom own dataset into COCO format'''
from detectron2.data import DatasetCatalog, MetadataCatalog

import random
from detectron2.utils.visualizer import Visualizer,ColorMode


from detectron2.data.datasets.coco import load_coco_json
```

```python
'''tnbc_patch'''

# [file path]
tnbc_root = 'data/cell-seg-data/tnbc_patch/'

# train
tnbc_imgs_train = os.path.join(tnbc_root, "img_train")
tnbc_ann_train = os.path.join(tnbc_root, "ann_train")
tnbc_outputJson_train = os.path.join(tnbc_root, "tnbc_train.json")

# val

# test
tnbc_test_root = "data/cell-seg-data/tnbc/"
# tnbc_test_root = 'data/cell-seg-data/tnbc_patch/'
tnbc_imgs_test = os.path.join(tnbc_test_root, "img_test")
tnbc_ann_test = os.path.join(tnbc_test_root, "ann_test")
tnbc_outputJson_test = os.path.join(tnbc_test_root, "tnbc_test.json")

# [register]
tnbc_train_set = "tnbc_train1"
DatasetCatalog.register(tnbc_train_set, lambda: load_coco_json(tnbc_outputJson_train, tnbc_imgs_train))
MetadataCatalog.get(tnbc_train_set).set(thing_classes=["background", "cell"])

tnbc_test_set = "tnbc_test1"
DatasetCatalog.register(tnbc_test_set, lambda: load_coco_json(tnbc_outputJson_test, tnbc_imgs_test))
MetadataCatalog.get(tnbc_test_set).set(thing_classes=["background", "cell"])

# [general]
train_set = tnbc_train_set
val_set = tnbc_train_set
test_set = tnbc_test_set

tag = "tnbc_patch"

input_min_size = 512
input_max_size = 512

dataset_dicts = load_coco_json(tnbc_outputJson_train, tnbc_imgs_train)
print(dataset_dicts)
```

```python
'''fluo_patch'''

# [file path]
fluo_root = 'data/cell-seg-data/fluo_patch/'

# train
fluo_imgs_train = os.path.join(fluo_root, "img_train")
fluo_ann_train = os.path.join(fluo_root, "ann_train")
fluo_outputJson_train = os.path.join(fluo_root, "fluo_train.json")

# val
fluo_imgs_val = os.path.join(fluo_root, "img_val")
fluo_ann_val = os.path.join(fluo_root, "ann_val")
fluo_outputJson_val = os.path.join(fluo_root, "fluo_val.json")

# test
# fluo_test_root = 'data/cell-seg-data/fluo_patch/'
fluo_test_root = 'data/cell-seg-data/fluo/'
fluo_imgs_test = os.path.join(fluo_test_root, "img_test")
fluo_ann_test = os.path.join(fluo_test_root, "ann_test")
fluo_outputJson_test = os.path.join(fluo_test_root, "fluo_test.json")

# [register]
fluo_train_set = "fluo_train"
DatasetCatalog.register(fluo_train_set, lambda: load_coco_json(fluo_outputJson_train, fluo_imgs_train))
MetadataCatalog.get(fluo_train_set).set(thing_classes=["background", "cell"])

fluo_val_set = "fluo_val"
DatasetCatalog.register(fluo_val_set, lambda: load_coco_json(fluo_outputJson_val, fluo_imgs_val))
MetadataCatalog.get(fluo_val_set).set(thing_classes=["background", "cell"])

fluo_test_set = "fluo_test"
DatasetCatalog.register(fluo_test_set, lambda: load_coco_json(fluo_outputJson_test, fluo_imgs_test))
MetadataCatalog.get(fluo_test_set).set(thing_classes=["background", "cell"])

# [general]
train_set = fluo_train_set
val_set = fluo_val_set 
test_set = fluo_test_set

tag = "fluo_patch"

input_min_size = 520
input_max_size = 696

dataset_dicts = load_coco_json(fluo_outputJson_test, fluo_imgs_test)
print(dataset_dicts)
```

```python
''' Visualize COCO JSON'''
def checkout_dataset_annotation(name):
    
    for i, d in enumerate(dataset_dicts,0):

        img = cv2.imread(d["file_name"])
        visualizer = Visualizer(img[:, :, ::-1], metadata=MetadataCatalog.get(name), scale=1)
        vis = visualizer.draw_dataset_dict(d)
        # cv2.imwrite("test.png", vis.get_image()[:, :, ::-1])
        cv2.imshow(d["file_name"], vis.get_image()[:, :, ::-1])

        if i == 2:
            break
    cv2.waitKey(0)

checkout_dataset_annotation(test_set)
```

```python
''' Train'''
from detectron2.utils.logger import setup_logger
from detectron2.config import get_cfg

# from detectron2.data import DatasetMapper, build_detection_train_loader
setup_logger()

cfg=get_cfg()

cfg.merge_from_file(model_zoo.get_config_file("Misc/mask_rcnn_R_50_FPN_3x_gn.yaml"))
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("Misc/mask_rcnn_R_50_FPN_3x_gn.yaml")

# cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_101_FPN_3x.yaml"))
# cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-InstanceSegmentation/mask_rcnn_R_101_FPN_3x.yaml")

cfg.DATASETS.TRAIN = (train_set,)
cfg.DATASETS.TEST = ()  # val set / no metrics implemented for this dataset?

cfg.TEST.EVAL_PERIOD = 100 #cfg.SOLVER.CHECKPOINT_PERIOD (?

cfg.DATALOADER.NUM_WORKERS = 1

cfg.SOLVER.IMS_PER_BATCH = 1
cfg.SOLVER.BASE_LR = 0.00020
cfg.SOLVER.MAX_ITER = 300
cfg.SOLVER.STEPS = []


cfg.MODEL.RESNETS.DEPTH = 101 
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 64
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 2
cfg.MODEL.DEVICE = "cuda"

cfg.OUTPUT_DIR = "output/"+tag

cfg.DATASETS.PRECOMPUTED_PROPOSAL_TOPK_TRAIN = 4000 #[]
cfg.DATASETS.PRECOMPUTED_PROPOSAL_TOPK_TEST = 2000 #[]

cfg.MODEL.RPN.POST_NMS_TOPK_TRAIN = 4000 #[]
cfg.MODEL.RPN.POST_NMS_TOPK_TEST = 2000 #[]

# train input size
cfg.INPUT.MIN_SIZE_TRAIN = 256
cfg.INPUT.MAX_SIZE_TRAIN = 256

cfg.TEST.DETECTIONS_PER_IMAGE = 500  #[]

os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)

from MyTrainer import MyTrainer

trainer = MyTrainer(cfg)
trainer.resume_or_load(resume=False)

# DefaultTrainer.test(cfg, trainer.model, evaluators=[COCOEvaluator(val_set, output_dir = cfg.OUTPUT_DIR)])

trainer.train()
```

```python
"""Inference"""
from detectron2.evaluation import COCOEvaluator
cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")  #    
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.7   # 

# test input size
cfg.INPUT.MIN_SIZE_TEST = input_min_size
cfg.INPUT.MAX_SIZE_TEST = input_max_size

# cfg.DATASETS.TEST = (test_set, )
predictor = DefaultPredictor(cfg)

"""Evaluation"""
from detectron2.evaluation import COCOEvaluator, inference_on_dataset
from detectron2.data import build_detection_test_loader

evaluator = COCOEvaluator(test_set, ("segm",), False, output_dir = cfg.OUTPUT_DIR)
val_loader = build_detection_test_loader(cfg, test_set)
print(inference_on_dataset(predictor.model, val_loader, evaluator))
```

```python
"""Visualization"""

for d in random.sample(dataset_dicts, 3):   # pick random 3
    im = cv2.imread(d["file_name"])
    outputs = predictor(im)
    v = Visualizer(im[:, :, ::-1],
                    metadata=MetadataCatalog.get(train_set),
                    scale=1,
                    instance_mode=ColorMode.IMAGE_BW   # remove the colors of unsegmented pixels
                    )
    v = v.draw_instance_predictions(outputs["instances"].to("cpu"))
    img = v.get_image()[:, :, ::-1]
    cv2.imshow(d["file_name"], img)
    cv2.waitKey(0)
```
